```python
from matplotlib import pyplot as plt
# import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Line Plot

```python
x = np.array([3, 5, 7, 9])
y = np.array([6, 10, 12, 15])
plt.plot(x,y, linewidth = '2.5', c = 'b', linestyle = '--')
plt.show()
```

### Title & Labels | Font Family

```python
x = np.array([3, 5, 7, 9])
y = np.array([6, 10, 12, 15])
plt.plot(x,y)

font1 = {'family':'calibri','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1, loc = 'center')
plt.xlabel("Average Pulse", fontdict = font2, loc = 'left')
plt.ylabel("Calorie Burnage", fontdict = font2, loc = 'bottom')
plt.show()


```

### Grids

```python
x = np.array([3, 5, 7, 9])
y = np.array([6, 10, 12, 15])
plt.plot(x,y)

font1 = {'family':'calibri','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1, loc = 'center')
plt.xlabel("Average Pulse", fontdict = font2, loc = 'left')
plt.ylabel("Calorie Burnage", fontdict = font2, loc = 'bottom')

# plt.grid()
# plt.grid(axis='x')
# plt.grid(axis='y')
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()


```

### Subplots

```python
font1 = {'family':'calibri','color':'blue','size':30}
font2 = {'family':'serif','color':'darkred','size':10}

x1 = np.array([3, 5, 7, 9])
y1 = np.array([6, 10, 12, 15])
plt.subplot(1,2,1)
plt.title("Employee", fontdict = font2, loc = 'center')
plt.xlabel("Average Pulse", fontdict = font2, loc = 'center')
plt.ylabel("Calorie Burnage", fontdict = font2, loc = 'center')
plt.plot(x1,y1)

x2 = np.array([3, 5, 7, 9])
y2 = np.array([6, 10, 12, 15])
plt.subplot(1,2,2)
plt.title("Customer", fontdict = font2, loc = 'center')
plt.xlabel("Average Pulse", fontdict = font2, loc = 'center')
plt.ylabel("Calorie Burnage", fontdict = font2, loc = 'center')
plt.plot(x2,y2)


plt.suptitle("Company", c="red", size=15)
plt.tight_layout()
plt.show()
```

### Bar Plot

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x,y, color='yellow', height = 0.1)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y, color = "red", width = 0.1)
plt.show()
```

### Pie Chart

```python
y = np.array([35.89, 25.78, 25, 15])
label = ["Delhi", "Mumbai", "Kolkata", "Chennai"]

myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = label, explode = myexplode, shadow = True)
plt.legend(loc="upper left")
plt.show()
```
